El dataset presentado reúne información académica correspondiente a tres generaciones de estudiantes (2017, 2018 y 2019) del Instituto Regional del Maule, ubicado en la comuna de San Javier. Los datos consideran el porcentaje de asistencia y el rendimiento escolar desde primero a septimo básico, organizados por asignatura y nivel.

Con el fin de proteger la identidad de los estudiantes, estos han sido identificados mediante su fecha de nacimiento, sin incluir nombres ni RUT.

Una característica importante del establecimiento es que, al finalizar sexto básico, realiza una clasificación de los estudiantes en dos niveles para cursar séptimo básico: Avanzado y No avanzado.

Para determinar el nivel de los estudiante se basan en el rendimiento académico obtenido en las asignaturas cientifico humanista de sexto básico.

La clasificacion se encuentra registrada en la columna denominada Nivel 7
